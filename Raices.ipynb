```python
import numpy as np
from scipy.optimize import newton
import matplotlib.pyplot as plt

def f(x):
    return x**5 + 5.5*x**4 + 4.5*x**3 - 9.5*x**2 - 15.5*x - 6

def f_prime(x):
    return 5*x**4 + 22*x**3 + 13.5*x**2 - 19*x - 15.5
```

```python
x_vals = np.linspace(-6, 4, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label='y = 0')
plt.title('Gráfica de la función f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.ylim(-50, 50)
plt.legend()
plt.show()
```

```python
p0_1 = -4.5
root_1 = newton(f, p0_1, fprime=f_prime)

p0_2 = -1.8
root_2 = newton(f, p0_2, fprime=f_prime)

p0_3 = 2.0
root_3 = newton(f, p0_3, fprime=f_prime)


all_roots = [root_1, root_2, root_3]

sorted_roots = sorted(all_roots, key=lambda x: (np.real(x), np.imag(x)))
```
