```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 7.5, "c": 0, "d": -3.5},
    {"a": 5, "b": -3, "c": -10.5, "d": 11.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red", label=f"$S_{i}(x)$" if i == 0 else "")

plt.scatter(xs, ys, zorder=5, label="Puntos originales")
x_tangent = np.linspace(-0.5, 0.5, 10)
y_tangent = -3 * x_tangent + 5
plt.plot(x_tangent, y_tangent, "b--", label="Pendiente deseada (m=-3)")
plt.text(0.1, 4, "$m_{deseada}$", fontsize=12)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolacin con Splines (Pendiente Modificada)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
